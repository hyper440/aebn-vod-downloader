```python
#@title üé¨ **Download** { display-mode: "form" }

#@markdown ### üîó Movie URL
movie_url = "" #@param {type:"string"}

# Early validation
if not movie_url:
    print("\n" + "="*50)
    print("‚ùå ERROR: Please enter a movie URL above!")
    print("="*50 + "\n")
    raise SystemExit("URL is required")

if "aebn.com" not in movie_url and "aebn.net" not in movie_url:
    print("\n" + "="*50)
    print("‚ùå ERROR: Invalid URL! Must be an AEBN URL.")
    print("   Example: https://straight.aebn.com/straight/movies/12345/title")
    print("="*50 + "\n")
    raise SystemExit("Invalid URL")

#@markdown ---
#@markdown ### üé• Video
resolution = 1080 #@param {type:"integer"}
force_resolution = False #@param {type:"boolean"}
target_stream = "both" #@param ["both", "video", "audio"]

#@markdown ---
#@markdown ### üé¨ Scene
split_scenes = False #@param {type:"boolean"}
scene_number = 0 #@param {type:"integer"}
#@markdown <small>Set scene_number > 0 to download single scene. Use split_scenes to download all scenes as separate files.</small>

#@markdown ---
#@markdown ### üìÅ Output
include_performer_names = False #@param {type:"boolean"}
download_covers = False #@param {type:"boolean"}
no_metadata = False #@param {type:"boolean"}

#@markdown ---
#@markdown ### ‚ö° Network
threads = 20 #@param {type:"slider", min:5, max:100, step:5}
proxy = "" #@param {type:"string"}
proxy_metadata_only = False #@param {type:"boolean"}

# Install
print("üì¶ Installing dependencies...")
!uv pip install https://github.com/hyper440/aebn-vod-downloader/archive/refs/heads/main.zip -q --system
print("‚úÖ Ready!\n")

# Build command
cmd = f"aebndl '{movie_url}' -r {resolution} -t {threads}"

if force_resolution:
    cmd += " -f"
if target_stream != "both":
    cmd += f" -ts {target_stream}"
if split_scenes:
    cmd += " --split-scenes"
elif scene_number > 0:
    cmd += f" -s {scene_number}"
if include_performer_names:
    cmd += " -n"
if download_covers:
    cmd += " -c"
if no_metadata:
    cmd += " -nm"
if proxy:
    cmd += f" -p '{proxy}'"
if proxy_metadata_only:
    cmd += " -pm"

print(f"üöÄ Running: {cmd}\n")
print("=" * 50)
!{cmd}
```

```python
#@title üíæ **Save to Computer** { display-mode: "form" }
#@markdown Click ‚ñ∂Ô∏è to download all output files to your computer.

import glob
from google.colab import files

mp4_files = glob.glob('*.mp4')
if not mp4_files:
    print("‚ùå No MP4 files found. Run the Download cell first.")
else:
    print(f"üìÅ Found {len(mp4_files)} file(s):\n")
    for mp4_file in mp4_files:
        print(f"  ‚¨áÔ∏è {mp4_file}")
        files.download(mp4_file)
    print("\n‚úÖ Done!")
```
